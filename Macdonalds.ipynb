```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

mcdonalds = pd.read_csv('C:/Users/gupta/Downloads/mcdonalds.csv')

MD_x = (mcdonalds.iloc[:, 0:11] == "Yes").astype(int)

scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(MD_x)

pca = PCA(n_components=11)
MD_pca = pca.fit(MD_x_scaled)

print("Standard deviations (1, .., p=11):")
print(np.round(MD_pca.explained_variance_ratio_**0.5, 1))

print("\nRotation (n x k) = (11 x 11):")
loadings = pd.DataFrame(MD_pca.components_.T, columns=[f"PC{i+1}" for i in range(11)], index=mcdonalds.columns[:11])
print(loadings.round(3))
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(mcdonalds)

pca = PCA(n_components=11)
MD_pca = pca.fit_transform(MD_x_scaled)


plt.figure(figsize=(8, 6))
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], c="grey", alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.grid(True)


loadings = pca.components_.T
for i, feature in enumerate(mcdonalds.columns):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='r', alpha=0.7, head_width=0.02)
    plt.text(loadings[i, 0] * 1.1, loadings[i, 1] * 1.1, feature, color='r', ha='center', va='center')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

np.random.seed(1234)

silhouette_scores = []
inertia_values = []
range_clusters = range(2, 9)

for k in range_clusters:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    labels = kmeans.fit_predict(MD_x)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(MD_x, labels))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range_clusters, inertia_values, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Inertia vs. Number of Clusters")

plt.subplot(1, 2, 2)
plt.plot(range_clusters, silhouette_scores, marker='o', color='orange')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")

plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np  # Add this import statement

mcdonalds = pd.read_csv('C:/Users/gupta/Downloads/mcdonalds.csv')

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

for column in mcdonalds.select_dtypes(include=['object']).columns:
    mcdonalds[column] = label.fit_transform(mcdonalds[column])

MD_x = mcdonalds.values

np.random.seed(1234)

nrep = 10
num_segments = range(1, 9)
within_cluster_distances = []
MD_km28 = {}

for k in num_segments:
    kmeans = KMeans(n_clusters=k, n_init=nrep, random_state=1234)
    kmeans.fit(MD_x)
    within_cluster_distances.append(kmeans.inertia_)
    MD_km28[str(k)] = kmeans

plt.bar(num_segments, within_cluster_distances)
plt.xlabel("Number of segments")
plt.ylabel("Sum of within-cluster distances")
plt.title("Segmentation Results")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

np.random.seed(1234)

gm_models = {}
bic_scores = []

for k in range(2, 9):
    gm = GaussianMixture(n_components=k, covariance_type='full', n_init=10, random_state=1234)
    gm.fit(MD_x)
    gm_models[k] = gm
    bic_scores.append(gm.bic(MD_x))

for k, model in gm_models.items():
    print(f"k={k}, BIC={model.bic(MD_x):.2f}")

gm_model_4 = gm_models[4]

kmeans = KMeans(n_clusters=4, n_init=10, random_state=1234)
kmeans_labels = kmeans.fit_predict(MD_x)

gm_labels = gm_model_4.predict(MD_x)

contingency_table = pd.crosstab(kmeans_labels, gm_labels, rownames=['KMeans'], colnames=['Gaussian Mixture'])
print(contingency_table)
```
